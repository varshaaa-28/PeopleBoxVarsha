```python
import pandas as pd
df=pd.read_csv('input.csv')
```

```python
df
```

```python
df=df.iloc[:3,:17]
df
```

```python
for i in df.select_dtypes(include=object).columns:
    df[i]=pd.to_datetime(df[i],format='%d-%m-%Y')
```

```python
df.info()
```

```python
da=[]
for i in range(3):
    dates=[]
    for j in df.iloc[i,]:
        if str(type(j))=="<class 'pandas._libs.tslibs.timestamps.Timestamp'>":
            dates.append(j)
    dates.sort()
    da.append(dates)
```

```python
da
```

```python
df1=pd.DataFrame(columns=['Employee Code', 'Manager Employee Code','Last Compensation','Compensation','Last Pay Raise Date','Performance Rating','Engagement Score','Effective Date','End Date'])
```

```python
df
```

```python
for i in range(len(da)):
    for j in range(len(da[i])):
        for k in df.iloc[i,].index:
            if da[i][j]==df.iloc[i,][k]:
                if k=='Date of Joining':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation'],da[i][j]]
                elif k=='Engagement 1 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation'],df.iloc[i,]['Engagement 1'],da[i][j]]
                elif k=='Review 1 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Performance Rating','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation'],df.iloc[i,]['Engagement 1'],df.iloc[i,]['Review 1'],da[i][j]]
                elif k=='Compensation 1 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Performance Rating','Last Compensation','Last Pay Raise Date','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation 1'],df.iloc[i,]['Engagement 1'],df.iloc[i,]['Review 1'],df.iloc[i,]['Compensation'],df.iloc[i,]['Compensation 1 date'],da[i][j]]
                elif k=='Engagement 2 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Performance Rating','Last Compensation','Last Pay Raise Date','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation 1'],df.iloc[i,]['Engagement 2'],df.iloc[i,]['Review 1'],df.iloc[i,]['Compensation'],df.iloc[i,]['Compensation 1 date'],da[i][j]]
                elif k=='Review 2 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Performance Rating','Last Compensation','Last Pay Raise Date','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation 1'],df.iloc[i,]['Engagement 2'],df.iloc[i,]['Review 2'],df.iloc[i,]['Compensation'],df.iloc[i,]['Compensation 1 date'],da[i][j]]
                elif k=='Compensation 2 date':
                    df1.loc[len(df1),['Employee Code','Manager Employee Code','Compensation','Engagement Score','Performance Rating','Last Compensation','Last Pay Raise Date','Effective Date']]=[i+1,df.iloc[i,]['Manager Employee Code'],df.iloc[i,]['Compensation 2'],df.iloc[i,]['Engagement 2'],df.iloc[i,]['Review 2'],df.iloc[i,]['Compensation 1'],df.iloc[i,]['Compensation 2 date'],da[i][j]]
                   
```

```python
from datetime import timedelta
```

```python
end_dates=[]
for i in range(len(da)):
    for j in range(len(da[i])):
        if j==len(da[i])-1:
            end_dates.append(pd.to_datetime('01-01-2100'))
        else:
            end_dates.append(da[i][j+1]-timedelta(days=1))
```

```python
end_dates.pop(15)
```

```python
df1['End Date']=end_dates
```
